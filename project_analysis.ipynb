### Motivation

Urban heat islands are a well known effect in cities, where urban areas tend to be warmer than surrounding rural areas. This is caused by changes in land cover, reduced vegetation, and more buildings and infrastructure that absorb incoming radiation. As climate change causes summer temperatures to increase across the United States, it is important to understand how cities are warming and if they are experiencing additional warming relative to rural areas. Many studies have documented urban head islands using station data or short term time periods; however, there are fewer studies that look into long term trends in the difference between urban and rural temperatures using spatial data. This project aims to assess the long term summer temperature trends across 5 major cities in the U.S., and to evaluate how urban heat island strength has changed relative to background climate warming in the past 75 years.

### Approach

This project uses monthly ERA5 2m temperature data during the summer months (JJA) from 1950 to 2024 across 5 major U.S. cities (New York, Chicago, Los Angeles, Atlanta, and Seattle). Urban regions were defined using a bounding box for each city, while surrounding rural regions were defined using an expanded buffer regions that excludes the urban core. Area-averaged summer temperatures were computed separately for urban and rural regions, and the Urban Heat Island Index (UHI) was calculated by finding the difference between these means. Long term trends were evaluated using linear regression in addition to using the 95th percentile thresholds based on a 1981â€“2010 baseline to contextualize extreme warm conditions. Additional analyses included early vs. late period comparisons and an exploratory analysis of the relationship between mean UHI strength and population density.

### Results

All five cities in this study (New York City, Chicago, Los Angeles, Atlanta, and Seattle) show statistically significant summer warming trends for both the urban and rural areas. This warming reflects the influence of large scale climate change. In all cases, the urban areas were found to be warmer than the surrounding rural areas during the summer, although the magnitude of the difference varies significantly by city with significant year-to-year variability. Long term trends in the UHI index are modest compared to the strong increasing trends in absolute temperature. Some cities, specifically New York and Atlanta, showed slightly larger and persistent UHI values, while cities such as Seattle and Los Angeles had a much weaker and more variable difference between urban and rural temperatures. This is most likely influenced by regional climate and proximity to water. The comparison between early and late periods found higher summer temperatures for both urban and rural areas across all cities in the late period. Changes in UHI between these two periods were generally minimal with only a modest increase in a few cities. Overall, these results suggest that while urbanization is contributing to localized summer warming, long term trends are driven by broader climate change. Finally, the comparison between UHI and population density found that more densely populated cities tend to exhibit stronger average UHI intensity. However, this relationship varies by city and is limited by the small sample size.

### Conclusions

This analysis ultimately showed that long-term summer warming across major U.S cities is primarily driven by regional climate change, while urban heat island effect provides a smaller secondary contribution. While urban areas remained consistently warmer than nearby rural areas, multidecadal changes in UHI strength are smaller and show more variability than changes in absolute temperature. These results highlight the importance of considering both large scale climate change and local urban characteristics when planning for future extreme heat risks.
